```
from pyspark.sql.functions import *
from pyspark.sql.types import *
from pyspark.sql.window import Window
```

```
sampleEmployee = spark.read.format('com.databricks.spark.csv').options(header='true',inferschema='true').load("/FileStore/tables/us_500.csv",header=True)
```

```
df = sampleEmployee.withColumn("number_of_time_replication", lit(100))
fn_for_arr = udf(lambda n : [n] * n, ArrayType(IntegerType()))
df = df.withColumn("number_of_time_replication", fn_for_arr(df.number_of_time_replication))
employeeDF = df.withColumn('number_of_time_replication', explode(df.number_of_time_replication)).drop('number_of_time_replication')
```

```
df3 = employeeDF.select("*").groupBy('city').count()
w = Window.orderBy(col('count').desc())
df3 = df3.withColumn("Dense_rank", dense_rank().over(w)).drop('count')
w = Window.orderBy(col('Dense_rank'))
CityEmployeeDensity = df3.withColumn("Sequence", row_number().over(w))
CityEmployeeDensity = CityEmployeeDensity.drop('Dense_rank')
```

```
VaccinationDrivePlan = employeeDF.join(CityEmployeeDensity, on = 'city', how = 'left')
VaccinationDrivePlan = VaccinationDrivePlan.orderBy(col('Sequence').asc(), col('city').asc())
```

```
display(VaccinationDrivePlan)
```

```
w = Window.partitionBy('Sequence').orderBy('Sequence')
df_temp = VaccinationDrivePlan.withColumn("Vaccination_Serial_Number", row_number().over(w))
df_final = df_temp.withColumn("Vaccination_Day", ceil(col('Vaccination_Serial_Number')/100))
df_final = df_final.groupBy('Sequence','city').agg(max("Vaccination_Day").alias("Days_for_Vaccination"))
```

```
display(df_final)
```

```
w = Window.orderBy('Sequence').rangeBetween(Window.unboundedPreceding, 0)
Final_Vaccination_Schedule = df_final.withColumn("In_Sequential_from_day_1", sum('Days_for_Vaccination').over(w))
Final_Vaccination_Schedule = Final_Vaccination_Schedule.withColumn("In_Parallel_from_day_1", col('Days_for_Vaccination'))
```

```
seq_days = Final_Vaccination_Schedule.agg(max('In_Sequential_from_day_1')).collect()[0][0]
par_days = Final_Vaccination_Schedule.agg(max('In_Parallel_from_day_1')).collect()[0][0]
print("Total days for Sequential Vaccination of one city at a time:", seq_days)
print("Total days for Parallel Vaccination across all cities:", par_days)
print("City wise days in both approach:")
display(Final_Vaccination_Schedule)
```
